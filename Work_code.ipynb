# 8. Ручна перевірка якості кластеризації

```python
def equality_finder(data):
    pairs_table = pd.DataFrame({'Param_1': [], 'Param_2': [], 'Correlation': []})
    pairs_table
    for i in range(0, data.shape[0]):
        for j in range(0, data.shape[1]):
            if (data.iloc[i, j] == 1):
                pair = pd.DataFrame({'Param_1': [data.index[i]], 'Param_2': [data.columns[j]], 'Correlation': [data.iloc[i, j]]})
                pairs_table = pairs_table.append(pair)
 
    return pairs_table
```

```python
table = equality_finder(kendall_corr)

for index, row in table.iterrows():
    if row.Param_1 != row.Param_2:
        print(row.Param_1, ' and ', row.Param_2, ' corr = ', row.Correlation)
```

```python
kendall_corr_1 = kendall_corr.copy()

def correlation_finder(data):
    pairs_table = pd.DataFrame({'Param_1': [], 'Param_2': [], 'Correlation': []})
    pairs_table
    for i in range(0, data.shape[0]):
        for j in range(0, data.shape[1]):
            if (data.iloc[i, j] > 0.8) and (data.iloc[i, j] < 1):
                pair = pd.DataFrame({'Param_1': [data.index[i]], 'Param_2': [data.columns[j]], 'Correlation': [data.iloc[i, j]]})
                pairs_table = pairs_table.append(pair)
 
    return pairs_table
```

```python
coef_table = correlation_finder(kendall_corr_1)
coef_table
```

```python
new_index = [*range(0, coef_table.shape[0], 1)]
coef_table["new_index"] = new_index
coef_table.set_index('new_index', inplace = True)
coef_table
```

```python
def drop_dooble(data):
    delete_table = pd.DataFrame({'Param_1': [], 'Param_2': [], 'Correlation': []})
    for i in range(0, data.shape[0] - 1):
        for j in range(i + 1, data.shape[0]):
            if (data.iloc[i,0] == data.iloc[j,1]) and (data.iloc[i,1] == data.iloc[j,0]) and (data.iloc[i,2] == data.iloc[j,2]):
                ind = data.loc[[j]].index[0]
                
                delete_table = delete_table.append(data.iloc[ind,:]) 

    return delete_table

del_tabl = drop_dooble(coef_table)
del_tabl
```

```python
coef_without_duplicates = pd.concat([coef_table, del_tabl]).drop_duplicates(keep=False)
coef_without_duplicates

coef_without_duplicates.to_csv("C:/Users/user/Desktop/Робота, заняття/Sense test/dataset/coef_without_duplicates.csv")
```

```python
import numpy as np

def calculate_outliers(data, colum):
    Q1 = data[colum].quantile(0.25)
    Q3 = data[colum].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
#    print(Q1, ', ', Q3, ', ', IQR, ', ', lower_limit, ', ', upper_limit)
    
    filter_data = data[(data[colum] >= lower_limit) & (data[colum] <= upper_limit)]
    return filter_data
    
calculate_outliers(out_set, 'group_1')
```

```python
def delete_outliers(data):
    temp_data = data
    for column in data:
        temp_data = calculate_outliers(temp_data, column)
        
    return(temp_data)
```
